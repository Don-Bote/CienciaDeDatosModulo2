```python
#Clase contacto
class Contacto:

  #Constructor contacto
  def __init__(self, nombre, telefono, correo, direccion):
    self.__nombre = nombre
    self.__telefono = telefono
    self.__correo = correo
    self.__direccion = direccion

  #Metodos Get
  def getNombre(self):
    return self.__nombre
  def getTelefono(self):
    return self.__telefono
  def getCorreo(self):
    return self.__correo
  def getDireccion(self):
    return self.__direccion

  #Metodos Set
  def setNombre(self, nombre):
    self.__nombre = nombre
  def setTelefono(self, telefono):
    self.__telefono = telefono
  def setCorreo(self, correo):
    self.__correo = correo
  def setDireccion(self, direccion):
    self.__direccion = direccion

  #Funcion para mostrar informacion con formato de diccionario
  def informacion(self):
    """Devuelve la informacion del contacto en forma de diccionario"""
    diccionario = {"nombre": self.__nombre,
                   "telefono": self.__telefono,
                   "correo": self.__correo,
                   "direccion": self.__direccion}
    return diccionario


```

```python
#funcion para escribir el menu
def mostrar_menu():
  print("-- Menu de opciones --")
  print("1. Agregar contacto")
  print("2. Modificar contactos")
  print("3. Borrar contacto")
  print("4. Buscar contacto")
  print("5. Mostrar todos los contactos")
  print("6. Salir")

#funcion para agregar un contacto (recibe una lista)
def agregar_contacto(lista):
  nombre = input("Ingrese el nombre del contacto: ")
  telefono = input("Ingrese el telefono del contacto: ")
  correo = input("Ingrese el correo del contacto: ")
  direccion = input("Ingrese la direccion del contacto: ")
  lista.append(Contacto(nombre, telefono, correo, direccion))

#funcion para modificar un contacto (recibe una lista)
def modificar_contacto(lista):
  print("\nModificar contacto por:")
  print("1. Nombre")
  print("2. Numero de telefono")
  print("3. Salir")
  opcion = int(input("Ingrese una opcion: "))
  if opcion == 1:
    posicion = buscar_por_nombre(lista)
    if posicion > -1:
      agregar_contacto(lista)
      lista.pop(posicion)
      print("\nContacto modificado\n")

  elif opcion == 2:
    posicion = buscar_por_numero(lista)
    if posicion > -1:
      agregar_contacto(lista)
      lista.pop(posicion)
      print("\nContacto modificado\n")

  elif opcion == 3:
    print("\nVolviendo al menu principal")
  else:
    print("\nOpcion no valida\nVolviendo al menu principal")

#funcion para borrar un contacto (recibe una lista)
def borrar_contacto(lista):
  print("\nBorrar contacto por:")
  print("1. Nombre")
  print("2. Numero de telefono")
  print("3. Salir")
  opcion = int(input("Ingrese una opcion: "))
  if opcion == 1:
    posicion = buscar_por_nombre(lista)
    if posicion > -1:
      lista.pop(posicion)
      print("\nContacto borrado\n")

  elif opcion == 2:
    posicion = buscar_por_numero(lista)
    if posicion > -1:
      lista.pop(posicion)
      print("\nContacto borrado\n")

  elif opcion == 3:
    print("\nVolviendo al menu principal")
  else:
    print("\nOpcion no valida\nVolviendo al menu principal")

#funcion para buscar un contacto recibe una lista
def buscar_contacto(lista):
  print("\nBuscar contacto por:")
  print("1. Nombre")
  print("2. Numero de telefono")
  print("3. Salir")
  opcion = int(input("Ingrese una opcion: "))
  if opcion == 1:
    posicion = buscar_por_nombre(lista)
    if posicion > -1:
      print("\nInformacion del contacto:")
      print(lista[posicion].informacion())
    return posicion

  elif opcion == 2:
    posicion = buscar_por_numero(lista)
    if posicion > -1:
      print("\nInformacion del contacto:")
      print(lista[posicion].informacion())
    return posicion

  elif opcion == 3:
    print("\nVolviendo al menu principal")
  else:
    print("\nOpcion no valida\nVolviendo al menu principal")

#funcion para buscar por nombre un contacto (recibe una lista y devuelve la pocicion en esa lista)
#este metodo es usado en los metodos anteiores
def buscar_por_nombre(lista):
  nombre = input("Ingrese el nombre del contacto: ")
  for i in range(len(lista)):
    if lista[i].getNombre() == nombre:
      print("contacto encontrado")
      return i
  print("\nContacto no encontrado")
  return -1

#funcion para buscar por numero de contacto (recibe una lista y devuelve la pocicion en esa lista)
#este metodo es usado en los metodos anteiores
def buscar_por_numero(lista):
  numero = input("Ingrese el numero del contacto: ")
  for i in range(len(lista)):
    if lista[i].getTelefono() == numero:
      print("contacto encontrado")
      return i
  print("\nContacto no encontrado")
  return -1

#funcion para mostrar todos los contactos de la lista
def mostrar_todos(lista):
  print("\nLista de contactos:")
  for i in range(len(lista)):
    print(lista[i].informacion())
```

```python
#variable menu para controlar el menu
menu = 0
#lista donde se guardan los contactos
lista_contactos = []
#while que controla el menu
while menu != 6:
  mostrar_menu()
  menu = int(input("Ingrese una opcion: "))

  #if para entrar a las opciones del menu y llamar a las funciones
  if menu == 1:
    agregar_contacto(lista_contactos)
    print("\nContacto agregado\n")

  elif menu == 2:
    modificar_contacto(lista_contactos)

  elif menu == 3:
    borrar_contacto(lista_contactos)

  elif menu == 4:
    posicion = buscar_contacto(lista_contactos)

  elif menu == 5:
    mostrar_todos(lista_contactos)

  elif menu == 6:
    print("\nSaliendo del programa")

  else:
    print("Opcion no valida")

  print("")
```
